```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
```

```python
data_X = np.array([[1], [2], [3]])  # This can have more features
data_Y = np.array([[3], [2], [4]])

model = linear_model.LinearRegression()
model.fit(data_X, data_Y)

predicted_Y = model.predict(data_X)

print(
    "Mean squared error is:  ",
    mean_squared_error(data_Y, predicted_Y),
)
# Our equation is y = mx + c
print("Weights:  ", model.coef_)
# In our case weight in only 1 (i.e. m)
print("Intercept:  ", model.intercept_)

print(f"Equation :- y = {round(model.coef_[0][0], 4)}x + {model.intercept_[0]}")
```

```python
plt.scatter(data_X, data_Y, label="Our actual data")
# Thiese  are the points we have

plt.plot(data_X, predicted_Y, label="Predicted Line")
# Actual line

plt.title("Basic Linear Regression")
plt.ylabel("House Price")
plt.xlabel("Area")
plt.legend()
plt.show()
```
